```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train_x = train.drop([
                      'date_time','target_carbon_monoxide','target_benzene','target_nitrogen_oxides'
],axis=1)
```

```python
labels = pd.DataFrame({
    'target_benzene':train['target_benzene'],
    'target_carbon_monoxide':train['target_carbon_monoxide'],
    'target_nitrogen_oxides':train['target_nitrogen_oxides']
})
```

```python
labels.describe()
```

```python
train_co = np.array(train['target_carbon_monoxide']).astype(np.float)
train_benzene = np.array(train['target_benzene']).astype(np.float) 
train_no = np.array(train['target_nitrogen_oxides']).astype(np.float)
```

```python
train_x
```

```python
test = test.drop('date_time',axis=1)
```

```python
test.head()
```

```python
model_co = tf.keras.models.Sequential([
                                       tf.keras.layers.Dense(64,activation='relu'),
                                       tf.keras.layers.Dense(1)
])

model_co.compile(
    optimizer = 'adam',
    loss = 'mse',
    metrics = ['mse']
)
```

```python
model_co.fit(train_x, train_co, epochs = 15)
```

```python
test_co = model_co.predict(test)
```

```python
model_benzene = tf.keras.models.Sequential([
                                       tf.keras.layers.Dense(64, activation='relu'),
                                       tf.keras.layers.Dense(1)
])

model_benzene.compile(
    optimizer = 'adam',
    loss = 'mse',
    metrics = ['mse']
)
```

```python
model_benzene.fit(train_x, train_benzene, epochs=150)
```

```python
test_benzene = model_benzene.predict(test)
```

```python
pd.DataFrame(train_benzene).describe()
```

```python
pd.DataFrame(test_benzene).describe()
```

```python
model_no = tf.keras.models.Sequential([
                                       tf.keras.layers.Dense(128, activation='relu'),
                                       tf.keras.layers.Dense(64, activation='relu'),
                                       tf.keras.layers.Dense(1, activation='relu')
])

model_no.compile(
    optimizer = 'adam',
    loss = 'mse',
    metrics = ['mse']
)
```

```python
model_no.fit(train_x, train_no, epochs = 15)
```

```python
test_no = model_no.predict(test)
```

```python
test1 = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test1['target_carbon_monoxide'] = test_co
test1['target_benzene'] = test_benzene
test1['target_nitrogen_oxides'] = test_no
```

```python
test1 = test1.drop(['deg_C','relative_humidity','absolute_humidity','sensor_1','sensor_2','sensor_3','sensor_4','sensor_5'],axis=1)
```

```python
test1
```

```python
test1.to_csv('submission.csv',index=False)
```
