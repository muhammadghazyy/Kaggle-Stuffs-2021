```python
import pycaret.regression
```

```python
train.drop('id',axis = 1,inplace=True)
```

## Tensorflow

```python
train1 = train.copy()
```

```python
train1.drop('id',axis=1,inplace = True)
```

```python
train1.drop('loss',axis = 1,inplace = True)
```

```python
target = train['loss']
```

```python
train1.head()
```

```python
train1.shape
```

```python
target.describe()
```

```python
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Dense(512),
                                    tf.keras.layers.Dropout(0.2),
                                    tf.keras.layers.Dense(1)
])

model.compile(
    optimizer = 'sgd',
    loss = 'mse',
    metrics = ['mse']
    )
```

```python
model.fit(train1,target,epochs = 2)
```

```python
test1 = test.copy()
test1.drop('id',axis=1,inplace = True)
```

```python
model.predict(test1[:1])
```
